## Text Pre-Processing Assignments

### 1. Using Pandas

```python
import pandas as pd
```

```python
# Prevent truncation of displayed text
pd.set_option('display.max_colwidth', None)
```

```python
books_df = pd.read_csv('../resources/Course Materials/Data/childrens_books.csv')
```

```python
print(books_df.shape)
books_df.head()
```

```python
# Convert Description data to lowercase
books_df.Description = books_df.Description.str.lower()
books_df.head()
```

```python
# Find rows where Description contains \xa0 (non-breaking space)
mask = books_df.Description.str.contains('\xa0', na=False)
rows_with_xa0 = books_df[mask]
print(f"{len(rows_with_xa0)} Rows containing \\xa0:")
```

```python
# Remove all \xa0 characters (non-breaking space) with normal space character....no need for regex
books_df.Description = books_df.Description.str.replace('\xa0', ' ')
books_df.head()
```

```python
# Find rows where Description contains \xa0 (non-breaking space)
mask = books_df.Description.str.contains('\xa0', na=False)
rows_with_xa0 = books_df[mask]
print(f"{len(rows_with_xa0)} Rows containing \\xa0:")
```

```python
# Remove all punctuation and trailing spaces
books_df.Description = books_df.Description.str.replace(r'[^\w\s]', '', regex=True)
books_df.Description = books_df.Description.str.replace(r'\s+$', '', regex=True)
books_df.head()
```

### 2. Using spaCy

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
# Provide a tokenized doc, so we can view its type and attributes
doc = nlp("For he's a jolly good fellow")
print(type(doc))
# print(vars(doc))   ....doesn't work because doc is not a pure python object and doesn't have __dict__)
# See just the main attributes (no methods or private ... or deprecated)
deprecated = {'is_nered', 'is_parsed', 'is_tagged', 'is_sentenced'}
doc_attrs = [attr for attr in dir(doc) 
             if not attr.startswith('_') 
             and attr not in deprecated
             and not callable(getattr(doc, attr))]
print(doc_attrs)
```

```python
# Get just the first token of the doc, so we can view its type and attributes
first_token = doc[0]
print(type(first_token))
# See just the main attributes (no methods or private)
token_attrs = [attr for attr in dir(first_token) 
               if not attr.startswith('_') 
               and not callable(getattr(first_token, attr))]
print(token_attrs)
```

```python
# put all text preprocessing steps into a function to better organize our code
def token_lemma_nonstop(text):
    # Convert text string to list of tokens
    doc = nlp(text)
    # The 'for token in doc' part of the list comprehension calls doc.__iter__()
    # This yields individual Token objects one at a time
    # Each Token object has attributes like lemma_, text, pos_, etc.
    output = [token.lemma_ for token in doc if not token.is_stop] # tokenize, lemmatize and remove stop words
    output = ' '.join(output) # convert list into string
    return output
```

```python
# token_lemma_nonstop takes a text string, so we use 'apply' to execute it for each item (string) of series (DF column)
books_df.Description = books_df.Description.apply(token_lemma_nonstop)
books_df.head()
```

### 3. Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
dtm = cv.fit_transform(books_df.Description)
dtm_df = pd.DataFrame(dtm.toarray(), columns=cv.get_feature_names_out())
dtm_df
```

```python
cv2 = CountVectorizer(stop_words='english', min_df=0.1)
dtm2 = cv2.fit_transform(books_df.Description)
dtm_df2 = pd.DataFrame(dtm2.toarray(), columns=cv2.get_feature_names_out())
dtm_df2
```

```python
# Top 10 terms
term_count = dtm_df2.sum() 
print(type(term_count))
top_terms = term_count.sort_values().tail(10)
top_terms
```

```python
# Bottom 10 terms
bottom_terms = term_count.sort_values().head(10)
bottom_terms
```

```python
# To get the largest value at the top of the plot, I had to use ascending sort and tail()
top_terms.plot(kind='barh');
```

### 4. TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer()
tfidf = tv.fit_transform(books_df.Description)
tfidf_df = pd.DataFrame(tfidf.toarray(), columns=tv.get_feature_names_out())
tfidf_df
```

```python
tv2 = TfidfVectorizer(stop_words='english', min_df=0.1, max_df=0.5)
tfidf2 = tv2.fit_transform(books_df.Description)
tfidf_df2 = pd.DataFrame(tfidf2.toarray(), columns=tv2.get_feature_names_out())
tfidf_df2
```

```python
# 10 most highly weighted words
weights = tfidf_df2.sum()
print(type(weights))
top_weights = weights.sort_values().tail(10)
top_weights
```

```python
top_weights.plot(kind='barh');
```

```python
import matplotlib.pyplot as plt

# Create side-by-side subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14,6))

top_terms.plot(kind='barh', ax=axes[0], color='blue')
axes[0].set_title('Count Vectorizer')

top_weights.plot(kind='barh', ax=axes[1], color='green')
axes[1].set_title('TF-IDF Vectorizer')

plt.tight_layout()
plt.show()
```
